```python
df_2103 = df.loc[(df['Year'] == 2021) & (df['Month'] == 3)]
```

```python
df_BE = df_2103.loc[df['EstimateType'] == 'BE'][['AocType', 'Novelty', 'AmountType', 'EconomicBasis', 'Value', 'LiabilityType']]
PVs = df_BE.set_index(['Novelty', 'AocType', 'EconomicBasis', 'LiabilityType'])['Value'].groupby(level=list(range(4))).sum().unstack(level=[-1, -2])
PVs
```

```python
df_RA = df_2103.loc[df['EstimateType'] == 'RA'][['AocType', 'Novelty', 'AmountType', 'EconomicBasis', 'Value', 'LiabilityType']]
RA = df_RA.set_index(['Novelty', 'AocType', 'EconomicBasis', 'LiabilityType'])['Value'].groupby(level=list(range(4))).sum().unstack(level=[-1, -2])
RA
```

```python
df_WA = df_2103.loc[df['EstimateType'] == 'A'][['AocType', 'Novelty', 'AmountType', 'Value', 'LiabilityType']]
WA = df_WA.set_index(['AocType', 'AmountType' , 'LiabilityType'])['Value'].groupby(level=list(range(3))).sum().unstack(level=[-1])
WA
```

```python
df_AAOA = df_2103.loc[(df['EstimateType'] == 'AA') | (df['EstimateType'] == 'OA')][['AocType', 'Novelty', 'AmountType', 'Value', 'EstimateType']]
AAOA = df_AAOA.set_index(['AocType', 'AmountType', 'EstimateType'])['Value'].groupby(level=list(range(3))).sum().unstack(level=[-1])
AAOA
```

```python
df_DA = df_2103.loc[df['EstimateType'] == 'DA'][['AocType', 'Novelty', 'AmountType', 'Value', 'EstimateType']]
DA = df_DA.set_index(['AocType', 'AmountType', 'EstimateType'])['Value'].groupby(level=list(range(3))).sum().unstack(level=[-1])
DA
```
